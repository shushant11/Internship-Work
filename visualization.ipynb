## 1.line plot

```python
import matplotlib.pyplot as plt

# sample data
months =['jan','feb','mar','april','may','jun']
energy_consumption = [1200,1300,1100,1500,1400,1600]

# plot a line plot
plt.plot(months,energy_consumption, marker='o', linestyle='--',color='b')
plt.xlabel('months')
plt.ylabel('energy_consumption (MWh)')
plt.title('energy consumption in 6 months')
plt.show()
```

## 2. bar chart

```python
# sample data
energy_sources = ['solar','wind','hydropower','biomass']
energy_values = [1200,3400,2900,2500]
# creating bar chart
plt.bar(energy_sources, energy_values, color='green')
# add title,axis (labels)
plt.title("energy consumption by renewable energy sources")
plt.xlabel("energy_source")
plt.ylabel("energy_consumption (MWh)")
plt.show()
```

## 3. pie chart 

```python
# pie chart to show energy consumption distribution
plt.pie(energy_values, labels= energy_sources , autopct='%1.1f%%',colors =['gold','lightblue','lightgreen','coral'])
plt.title('energy consumption share by source')
plt.show()
```

## 4. scatter plot

```python
# sample data for energy consumption and carbon emission
carbon_emission = [200,500,450,300] # in kg co2
# create a scatter plot
plt.scatter(energy_values,carbon_emission, color='b')
# add titles and labels
plt.title('energy consumption vs carbon emission')
plt.xlabel('energy consumption in MWh')
plt.ylabel('carbon emission in kg co2')
plt.show()
```

## 5.customization

```python
# create a bar chart for customization
plt.bar(energy_sources , energy_values, color='orange',edgecolor='black',linewidth=2)
# add titles and labels
plt.title('customized energy consumption by source', fontsize=14 , fontweight='bold')
plt.xlabel('energy source', fontsize=12)
plt.ylabel('energy consumption in MWh', fontsize=12)
# add grid and style
plt.grid(True,linestyle='--', alpha=0.7)
plt.legend(['energy consumption'] ,loc='upper left')
plt.show()
```
